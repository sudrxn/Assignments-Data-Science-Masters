```python
#ans4

Bokeh server is a component of the Bokeh library that allows you to create interactive and dynamic web
applications with real-time updates. It enables you to build complex and interactive visualizations by
adding interactivity to your plots and charts. 
```

```python
# Create a Bokeh App:

# Filename: myapp.py

from bokeh.plotting import figure, curdoc
from bokeh.models import Slider
from bokeh.layouts import column

# Create a simple plot
plot = figure(width=400, height=300, title="Dynamic Plot")
line = plot.line(x=[1, 2, 3], y=[4, 5, 2])

# Create a slider widget
slider = Slider(start=1, end=10, value=1, step=1, title="Multiplier")

# Define callback function for slider
def update(attr, old, new):
    line.data_source.data['y'] = [4, 5, 2] * slider.value

# Attach callback to the slider
slider.on_change('value', update)

# Arrange the plot and slider in a layout
layout = column(plot, slider)

# Add the layout to the current document
curdoc().add_root(layout)




# Run the Bokeh Server:

bokeh serve myapp.py
```

```python
#ans5

Embedding a Bokeh plot into a web page or dashboard using Flask or Django involves creating the Bokeh plot,
generating the necessary HTML and JavaScript code, and then integrating it into your web framework of
choice
```

```python
pip install flask bokeh

#Create a Flask app:

# app.py
from flask import Flask, render_template
from bokeh.plotting import figure
from bokeh.embed import components

app = Flask(__name__)

@app.route('/')
def index():
    # Create Bokeh plot
    plot = figure(plot_width=400, plot_height=400)
    plot.circle([1, 2, 3, 4, 5], [6, 7, 2, 4, 5])

    # Embed plot into HTML
    script, div = components(plot)
    return render_template('index.html', script=script, div=div)

if __name__ == '__main__':
    app.run(debug=True)

   

 #Create a template (index.html):
    
    
    
```
