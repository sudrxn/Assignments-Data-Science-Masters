# ans1:

The mathematical formulation for a linear Support Vector Machine (SVM) can be expressed as an optimization problem. Given a set of training data with input vectors \(x_i\) and corresponding binary class labels \(y_i\), where \(y_i\) can be either -1 or 1, a linear SVM seeks to find a hyperplane defined by the weights \(w\) and bias \(b\) that maximally separates the data into two classes. The decision function for a linear SVM is given by:

\[ f(x) = \text{sign}(\mathbf{w} \cdot \mathbf{x} + b) \]

Here:
- \( \mathbf{w} \) is the weight vector.
- \( \mathbf{x} \) is the input vector.
- \( b \) is the bias term.
- \( \cdot \) represents the dot product.

The optimization problem for finding \( \mathbf{w} \) and \( b \) involves minimizing the following objective function, subject to certain constraints:

\[ \min_{\mathbf{w}, b} \frac{1}{2} \|\mathbf{w}\|^2 \]

Subject to the constraints:

\[ y_i(\mathbf{w} \cdot \mathbf{x}_i + b) \geq 1 \text{ for all } i \]

These constraints ensure that each data point is correctly classified with a margin of at least 1. The optimization is typically solved using techniques such as quadratic programming.

It's important to note that SVMs can be extended to non-linear cases using techniques like the kernel trick, but the above formula specifically represents the linear SVM case.

# ans2:

The objective function of a linear Support Vector Machine (SVM) is to find the hyperplane that maximally separates the data into different classes. The SVM is a supervised machine learning algorithm used for classification and regression tasks.

For linearly separable data, the objective function of a linear SVM can be expressed as:

\[ \text{minimize} \left( \frac{1}{2} ||w||^2 \right) \]

subject to the constraints:

\[ y_i (w \cdot x_i + b) \geq 1 \]

Here:
- \(w\) is the weight vector that defines the hyperplane.
- \(b\) is the bias term.
- \(x_i\) represents the feature vector of the \(i\)-th data point.
- \(y_i\) is the class label of the \(i\)-th data point, which is either +1 or -1 for binary classification.

The objective is to minimize the norm of the weight vector (\(||w||\)) while ensuring that all data points are correctly classified with a margin of at least 1. The margin is the distance between the hyperplane and the nearest data point from either class.

The constraints state that each data point should lie on the correct side of the decision boundary, with a margin of at least 1. The optimization problem aims to find the optimal values of \(w\) and \(b\) that satisfy these constraints and maximize the margin, resulting in an effective separation between classes.

# ans3:

The kernel trick is a key concept in Support Vector Machines (SVMs) that allows them to efficiently handle non-linearly separable data by implicitly mapping the input features into a higher-dimensional space. SVMs work well when the data is linearly separable, but in real-world scenarios, many datasets are not linearly separable. The kernel trick helps SVMs address this limitation.

In the context of SVM, a kernel is a function that computes the dot product between two transformed feature vectors in a higher-dimensional space without explicitly calculating the transformation. The SVM algorithm primarily relies on the dot product of data points, and the kernel trick allows it to operate in a high-dimensional space without explicitly computing the transformation, which can be computationally expensive.

The most commonly used kernels in SVM include:

1. **Linear Kernel (no kernel):** This is equivalent to a linear SVM, and it works well when the data is linearly separable.

2. **Polynomial Kernel:** It introduces non-linearity by computing the dot product raised to a certain power, controlled by a user-specified parameter.

3. **Radial Basis Function (RBF) Kernel:** Also known as the Gaussian kernel, it introduces non-linearity by transforming the data into an infinite-dimensional space using a Gaussian function.

4. **Sigmoid Kernel:** It uses a hyperbolic tangent function to introduce non-linearity.

The choice of kernel depends on the nature of the data and the problem at hand. The kernel trick allows SVMs to effectively handle non-linear relationships in the data by implicitly operating in a higher-dimensional space without explicitly computing the transformation, making SVMs versatile for a wide range of applications.

# asn 4:

Support vectors play a crucial role in Support Vector Machines (SVM), a supervised machine learning algorithm used for classification and regression tasks. In SVM, the algorithm aims to find a hyperplane that best separates the data into different classes. Support vectors are the data points that lie closest to the decision boundary (hyperplane) and have a direct influence on the positioning and orientation of the hyperplane.

Here's an explanation with an example:

### Example:

Let's consider a binary classification problem where we have two classes, positive (1) and negative (-1), and we want to find a hyperplane to separate these classes.

#### Data Points:
- Positive class (1): A, B
- Negative class (-1): C, D, E

#### Initial Scenario:
We have a 2D feature space, and initially, our data points and a potential decision boundary (hyperplane) might look like this:

```
        |        C (-1)
        |
        |   D (-1)
        |
--------|-------------------
        |
        |   A (1)
        |
        |   B (1)
```

#### Hyperplane:
The hyperplane is represented as a line, and SVM aims to find the optimal hyperplane that maximizes the margin, which is the distance between the hyperplane and the nearest data points (support vectors).

```
        |        C (-1)
        |
        |   D (-1)
        |
--------|---------(Optimal Hyperplane)-----
        |
        |   A (1)
        |
        |   B (1)
```

#### Support Vectors:
In this case, the support vectors are the points A, C, and D, as they are the ones closest to the hyperplane. These support vectors determine the position and orientation of the hyperplane. If any other points were chosen as support vectors, the hyperplane might not be optimal in terms of maximizing the margin.

### Role of Support Vectors:

1. **Determining the Hyperplane:** The position and orientation of the hyperplane are determined by the support vectors.

2. **Margin Maximization:** SVM aims to maximize the margin between the two classes, and the margin is defined by the distance between the hyperplane and the closest support vectors.

3. **Robustness:** SVM is robust to outliers because it mainly depends on the support vectors, and outliers that are not support vectors have less influence on the model.

In summary, support vectors are critical in SVM as they are the key data points that determine the optimal hyperplane, ensuring effective separation between different classes while maximizing the margin.

# asn 5:

Support Vector Machines (SVM) are a class of supervised learning algorithms that can be used for classification or regression tasks. SVM aims to find a hyperplane that best separates the data into different classes. Let's discuss Hyperplane, Marginal Plane, Soft Margin, and Hard Margin in SVM with examples and graphs:

1. **Hyperplane:**
   - A hyperplane is a decision boundary that separates the data points of different classes in feature space.
   - In a two-dimensional space, a hyperplane is a line; in three dimensions, it's a plane, and in higher dimensions, it's a hyperplane.
   - Example: Consider a 2D space with two classes (blue and red). A hyperplane can be a line that best separates the blue and red points.

   ![Hyperplane](images/svm_hyperplane.png)

2. **Margin:**
   - The margin is the distance between the hyperplane and the nearest data point of any class.
   - A larger margin indicates a more confident classification.
   - In SVM, the goal is to maximize the margin.

   ![Margin](images/svm_margin.png)

3. **Hard Margin:**
   - In Hard Margin SVM, the goal is to find a hyperplane that perfectly separates the data with no misclassifications.
   - It works well when the data is linearly separable.
   - Example: In a 2D space, a hard margin hyperplane perfectly separates blue and red points with no points in the margin.

   ![Hard Margin](images/svm_hard_margin.png)

4. **Soft Margin:**
   - In real-world scenarios, data may not be perfectly separable, or outliers might be present. Soft Margin SVM allows for some misclassifications.
   - It introduces a penalty for misclassifications, balancing the trade-off between maximizing the margin and allowing some errors.
   - Example: In the presence of outliers or noise, a soft margin hyperplane allows for a few misclassifications to achieve a better overall performance.

   ![Soft Margin](images/svm_soft_margin.png)

These graphs illustrate the concepts of a hyperplane, margin, hard margin, and soft margin in SVM. The choice between hard and soft margins depends on the nature of the data and the tolerance for misclassifications in the application.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

# Sample data
X = np.array([[1, 2], [2, 3], [3, 3], [2, 1], [3, 2]])
y = np.array([1, 1, 1, -1, -1])

# Create a SVM model with a linear kernel
clf = svm.SVC(kernel='linear', C=1.0)

# Fit the model
clf.fit(X, y)

# Plot the data points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, marker='o', s=100)

# Plot the hyperplane
w = clf.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(1, 3)
yy = a * xx - (clf.intercept_[0]) / w[1]
plt.plot(xx, yy, 'k-')

# Plot the margin
margin = 1 / np.sqrt(np.sum(clf.coef_ ** 2))
yy_down = yy - np.sqrt(1 + a ** 2) * margin
yy_up = yy + np.sqrt(1 + a ** 2) * margin
plt.plot(xx, yy_down, 'k--')
plt.plot(xx, yy_up, 'k--')

# Highlight support vectors
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1],
            s=200, facecolors='none', edgecolors='k')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('SVM Hyperplane, Margin, Hard Margin, and Soft Margin')

plt.show()
````

````python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

class LinearSVM:
    def __init__(self, learning_rate=0.01, epochs=1000, C=1.0):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.C = C
        self.weights = None
        self.bias = None

    def hinge_loss(self, X, y, weights, bias):
        margin = 1 - y * (np.dot(X, weights) + bias)
        margin[margin < 0] = 0
        loss = np.sum(margin) / X.shape[0] + 0.5 * self.C * np.sum(weights ** 2)
        return loss

    def fit(self, X, y):
        num_samples, num_features = X.shape
        self.weights = np.zeros(num_features)
        self.bias = 0

        for epoch in range(self.epochs):
            for i in range(num_samples):
                condition = y[i] * (np.dot(X[i], self.weights) + self.bias) >= 1
                gradient_weights = self.C * self.weights if not condition else 0
                gradient_bias = 0 if not condition else 0
                gradient_weights += -y[i] * X[i]
                gradient_bias += -y[i]

                self.weights -= self.learning_rate * gradient_weights
                self.bias -= self.learning_rate * gradient_bias

    def predict(self, X):
        return np.sign(np.dot(X, self.weights) + self.bias)

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = np.where(iris.target == 0, -1, 1)  # Convert labels to -1 and 1 for binary classification

# Choose two features for visualization
feature_index1 = 0
feature_index2 = 1
X_subset = X[:, [feature_index1, feature_index2]]

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_subset, y, test_size=0.2, random_state=42)

# Train linear SVM from scratch
linear_svm_scratch = LinearSVM(learning_rate=0.01, epochs=1000, C=1.0)
linear_svm_scratch.fit(X_train, y_train)

# Predict labels for the testing set
y_pred_scratch = linear_svm_scratch.predict(X_test)

# Calculate accuracy for the scratch implementation
accuracy_scratch = accuracy_score(y_test, y_pred_scratch)
print(f"Accuracy for Linear SVM (from scratch): {accuracy_scratch * 100:.2f}%")

# Train linear SVM using scikit-learn
from sklearn.svm import SVC

linear_svm_sklearn = SVC(kernel='linear', C=1.0)
linear_svm_sklearn.fit(X_train, y_train)

# Predict labels for the testing set using scikit-learn
y_pred_sklearn = linear_svm_sklearn.predict(X_test)

# Calculate accuracy for the scikit-learn implementation
accuracy_sklearn = accuracy_score(y_test, y_pred_sklearn)
print(f"Accuracy for Linear SVM (scikit-learn): {accuracy_sklearn * 100:.2f}%")
````
