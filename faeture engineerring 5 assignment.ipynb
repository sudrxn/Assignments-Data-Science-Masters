The Pearson correlation coefficient, often denoted as \( r \), measures the strength and direction of a linear relationship between two variables. It ranges from -1 to 1, where:

- \( r = 1 \) indicates a perfect positive linear relationship.
- \( r = -1 \) indicates a perfect negative linear relationship.
- \( r = 0 \) indicates no linear relationship.

To calculate the Pearson correlation coefficient between the amount of time students spend studying for an exam and their final exam scores, you would follow these steps:

1. Organize your data into pairs of corresponding values (time spent studying, exam scores).
2. Calculate the mean (\( \bar{x} \)) and standard deviation (\( s_x \)) of the time spent studying and exam scores.
3. Use the formula for Pearson correlation coefficient:

\[ r = \frac{\sum{(x_i - \bar{x})(y_i - \bar{y})}}{n \cdot s_x \cdot s_y} \]

where \( x_i \) and \( y_i \) are the individual data points, \( \bar{x} \) and \( \bar{y} \) are the means, \( s_x \) and \( s_y \) are the standard deviations, and \( n \) is the number of data points.

Once you've calculated \( r \), you can interpret the result:

- If \( r \) is close to 1, there is a strong positive linear relationship: as the time spent studying increases, exam scores tend to increase.
- If \( r \) is close to -1, there is a strong negative linear relationship: as the time spent studying increases, exam scores tend to decrease.
- If \( r \) is close to 0, there is a weak or no linear relationship between the two variables.

Remember that correlation does not imply causation. Even if there is a strong correlation, it doesn't necessarily mean that studying more causes higher exam scores; other factors could be involved.

To calculate Spearman's rank correlation, follow these steps:

1. Rank the data for each variable separately.
2. Calculate the difference between the ranks for each pair of corresponding data points.
3. Square the differences.
4. Sum up the squared differences.
5. Use the formula for Spearman's rank correlation:

\[ \rho = 1 - \frac{6 \sum d_i^2}{n(n^2 - 1)} \]

where:
- \( \rho \) is the Spearman's rank correlation coefficient,
- \( d_i \) is the difference between the ranks of corresponding data points,
- \( n \) is the number of data points.

Let's say you have the following data:

| Individual | Sleep (X) | Satisfaction (Y) |
|------------|-----------|-------------------|
| 1          | 7         | 8                 |
| 2          | 6         | 6                 |
| 3          | 8         | 9                 |
| 4          | 5         | 4                 |
| 5          | 7         | 7                 |

1. Rank the data for each variable:

| Individual | Sleep (X) | Rank(X) | Satisfaction (Y) | Rank(Y) |
|------------|-----------|---------|-------------------|---------|
| 1          | 7         | 3       | 8                 | 2       |
| 2          | 6         | 4       | 6                 | 5       |
| 3          | 8         | 2       | 9                 | 1       |
| 4          | 5         | 5       | 4                 | 4       |
| 5          | 7         | 3       | 7                 | 3       |

2. Calculate the differences (d) between the ranks for each pair: \( d_i = \text{Rank}(X)_i - \text{Rank}(Y)_i \)

| Individual | \( d_i \) |
|------------|----------|
| 1          | 1        |
| 2          | -1       |
| 3          | 1        |
| 4          | 1        |
| 5          | 0        |

3. Square the differences: \( d_i^2 \)

| Individual | \( d_i^2 \) |
|------------|------------|
| 1          | 1          |
| 2          | 1          |
| 3          | 1          |
| 4          | 1          |
| 5          | 0          |

4. Sum up the squared differences: \( \sum d_i^2 = 4 \)

5. Use the formula:

\[ \rho = 1 - \frac{6 \sum d_i^2}{n(n^2 - 1)} \]

\[ \rho = 1 - \frac{6 \times 4}{5 \times (5^2 - 1)} \]

\[ \rho = 1 - \frac{24}{100} \]

\[ \rho = 0.76 \]

Interpretation: The Spearman's rank correlation coefficient (\( \rho \)) is 0.76, indicating a strong positive monotonic relationship between the amount of sleep individuals get each night and their overall job satisfaction level. This suggests that as the amount of sleep increases, job satisfaction tends to also increase, and vice versa.

To calculate the Pearson correlation coefficient and Spearman's rank correlation coefficient, you'll need the data for the number of hours of exercise per week and the corresponding body mass index (BMI) for each participant. Let's assume you have the following hypothetical data:

```
Hours of Exercise (X):  [10, 8, 12, 15, 7, 6, 11, 9, 14, 13, 16, 8, 10, 12, 11, 9, 14, 13, 15, 7, 6, 12, 10, 14, 13, 16, 8, 10, 12, 11, 9, 14, 13, 15, 7, 6, 12, 10, 14, 13, 16, 8, 10, 12, 11, 9, 14, 13, 15]
BMI (Y):               [25, 27, 23, 21, 28, 29, 24, 26, 22, 23, 20, 27, 25, 23, 24, 26, 22, 23, 21, 28, 29, 23, 25, 22, 23, 20, 27, 25, 23, 24, 26, 22, 23, 21, 28, 29, 23, 25, 22, 23, 20, 27, 25, 23, 24, 26, 22, 23, 21]
```

Now, you can calculate the Pearson correlation coefficient and Spearman's rank correlation coefficient.

### Pearson Correlation Coefficient (r):

The formula for the Pearson correlation coefficient is:

\[ r = \frac{\sum{(X_i - \bar{X})(Y_i - \bar{Y})}}{\sqrt{\sum{(X_i - \bar{X})^2} \cdot \sum{(Y_i - \bar{Y})^2}}} \]

Where:
- \(X_i\) is the \(i\)-th value of the variable X (hours of exercise),
- \(Y_i\) is the \(i\)-th value of the variable Y (BMI),
- \(\bar{X}\) is the mean of X,
- \(\bar{Y}\) is the mean of Y.

Calculate \(r\) using the provided data:

```python
import numpy as np

X = np.array([10, 8, 12, 15, 7, 6, 11, 9, 14, 13, 16, 8, 10, 12, 11, 9, 14, 13, 15, 7, 6, 12, 10, 14, 13, 16, 8, 10, 12, 11, 9, 14, 13, 15, 7, 6, 12, 10, 14, 13, 16, 8, 10, 12, 11, 9, 14, 13, 15])
Y = np.array([25, 27, 23, 21, 28, 29, 24, 26, 22, 23, 20, 27, 25, 23, 24, 26, 22, 23, 21, 28, 29, 23, 25, 22, 23, 20, 27, 25, 23, 24, 26, 22, 23, 21, 28, 29, 23, 25, 22, 23, 20, 27, 25, 23, 24, 26, 22, 23, 21])

# Calculate means
mean_X = np.mean(X)
mean_Y = np.mean(Y)

# Calculate Pearson correlation coefficient
numerator = np.sum((X - mean_X) * (Y - mean_Y))
denominator = np.sqrt(np.sum((X - mean_X)**2) * np.sum((Y - mean_Y)**2))
pearson_corr_coeff = numerator / denominator

print("Pearson Correlation Coefficient (r):", pearson_corr_coeff)
```

### Spearman's Rank Correlation Coefficient (\(\rho\)):

The formula for Spearman's rank correlation coefficient is:

\[ \rho = 1 - \frac{6 \sum{d_i^2}}{n(n^2 - 1)} \]

Where:
- \(d_i\) is the difference between the ranks of corresponding pairs of values,
- \(n\) is the number of pairs.

Calculate \(\rho\) using the provided data:

```python
from scipy.stats import spearmanr

# Calculate Spearman's rank correlation coefficient
spearman_corr_coeff, _ = spearmanr(X, Y)

print("Spearman's Rank Correlation Coefficient (œÅ):", spearman_corr_coeff)
```

````python
#asn5:

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a DataFrame with the provided data
data = {'Age': [25, 42, 37, 19, 31, 28],
        'Brand': ['Coke', 'Pepsi', 'Mountain Dew', 'Coke', 'Pepsi', 'Coke']}
df = pd.DataFrame(data)

# Calculate average age for each brand
average_age_by_brand = df.groupby('Brand')['Age'].mean()

# Visualization: Bar chart
sns.barplot(x='Brand', y='Age', data=df, estimator=lambda x: sum(x) / len(x), ci=None)
plt.title('Average Age by Brand Preference')
plt.show()

# Visualization: Count plot
sns.countplot(x='Brand', data=df)
plt.title('Brand Preference Count')
plt.show()

# Correlation Analysis: Convert categorical variables to numerical using Label Encoding
df['Brand'] = df['Brand'].astype('category').cat.codes

# Calculate correlation
correlation_matrix = df.corr()

# Visualization: Heatmap of the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()

````

````python
#ans6:

import numpy as np

# Replace these arrays with your actual data
sales_calls_per_day = np.array([30, 35, 40, 25, 28, 32, 37, 42, 29, 31, 38, 36, 27, 33, 39, 26, 34, 41, 24, 30, 35, 40, 25, 28, 32, 37, 42, 29, 31, 38])
sales_per_week = np.array([5, 8, 4, 5, 7, 9, 10, 6, 5, 8, 7, 4, 6, 9, 3, 7, 8, 4, 5, 6, 8, 4, 5, 7, 9, 10, 6, 5, 8, 7])

# Calculate means
mean_sales_calls_per_day = np.mean(sales_calls_per_day)
mean_sales_per_week = np.mean(sales_per_week)

# Calculate deviations from the mean
deviations_x = sales_calls_per_day - mean_sales_calls_per_day
deviations_y = sales_per_week - mean_sales_per_week

# Calculate the Pearson correlation coefficient
numerator = np.sum(deviations_x * deviations_y)
denominator = np.sqrt(np.sum(deviations_x**2) * np.sum(deviations_y**2))
correlation_coefficient = numerator / denominator

print("Pearson correlation coefficient:", correlation_coefficient)

````
