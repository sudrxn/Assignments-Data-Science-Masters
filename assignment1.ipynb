```python
 
   Q1: Explain with an example each when to use a for loop and a while loop.

  ans:  The choice between using a for loop and a while loop depends on the specific requirements of the task that you're trying to accomplish.
        Here's an example of when to use a for loop:
        Suppose you have a list of numbers and you want to print each number in the list. Here's how you could use a for loop to accomplish this:
        
        numbers = [1, 2, 3, 4, 5]
        for num in numbers:
            print(num)
 
   
       
    
        In this example, the for loop will iterate over each number in the list, and the block of code inside the loop will be executed once for each iteration.
        The variable num will take on the value of each element in the list, one at a time.
        
        
        Here's an example of when to use a while loop: Suppose you want to print the numbers from 1 to 5.
        Here's how you could use a while loop to accomplish this:
        
        
        num = 1
        while num <= 5:
            print(num)
            num += 1

       
         In this example, the while loop will continue to execute as long as the condition num <= 5 is true.
        The code inside the loop will increase the value of num by 1 on each iteration, 
        so eventually the condition will become false and the loop will terminate
        
        
        
        
        
   Q2. Write a python program to print the sum and product of the first 10 natural numbers using for
       and while loop.     
        
    ans: Here is a Python program that prints the sum and product of the first 10 natural numbers using both a for loop and a while loop:
        
     # For Loop
sum = 0
product = 1
for i in range(1, 11):
    sum += i
    product *= i

print("Sum using for loop:", sum)
print("Product using for loop:", product)

# While Loop
sum = 0
product = 1
i = 1
while i <= 10:
    sum += i
    product *= i
    i += 1

print("Sum using while loop:", sum)
print("Product using while loop:", product)

        
        Sum using for loop: 55
        Product using for loop: 3628800
        Sum using while loop: 55
        Product using while loop: 3628800

        
        
  Q3. Create a python program to compute the electricity bill for a household.The per-unit charges in rupees are as follows: 
      For the first 100 units, the user will be charged Rs. 4.5 per
      unit, for the next 100 units, the user will be charged Rs. 6 per unit, and for the next 100 units, the user will
      be charged Rs. 10 per unit, After 300 units and above the user will be charged Rs. 20 per unit.
      You are required to take the units of electricity consumed in a month from the user as input.
      Your program must pass this test case: when the unit of electricity consumed by the user in a month is
      310, the total electricity bill should be 2250.
  ans:#Program to calculate electricity bill

units = int(input("Enter the number of units consumed: "))

#Calculating electricity bill for the given conditions
if units <= 100:
  bill = units * 4.5
elif units <= 200:
  bill = (100 * 4.5) + ((units-100) * 6)
elif units <= 300:
  bill = (100 * 4.5) + (100 * 6) + ((units-200) * 10)
else:
  bill = (100 * 4.5) + (100 * 6) + (100 * 10) + ((units-300) * 20)

print("The total electricity bill is:", bill)




     Q4. Create a list of numbers from 1 to 100. Use for loop and while loop to calculate the cube of each
         number and if the cube of that number is divisible by 4 or 5 then append that number in a list and print
         that list.
       
    ans:   # Using for loop
for_loop_result = []
for i in range(1, 101):
    cube = i**3
    if cube % 4 == 0 or cube % 5 == 0:
        for_loop_result.append(i)
print("Numbers whose cube is divisible by 4 or 5 (using for loop):", for_loop_result)

# Using while loop
i = 1
while_loop_result = []
while i <= 100:
    cube = i**3
    if cube % 4 == 0 or cube % 5 == 0:
        while_loop_result.append(i)
    i += 1
print("Numbers whose cube is divisible by 4 or 5 (using while loop):", while_loop_result.)



     Q5. Write a program to filter count vowels in the below-given string.
         string = "I want to become a data scientist"

    ans:  
        string = "I want to become a data scientist"
        vowels = "aeiouAEIOU"
        count = 0

        for char in string:
            if char in vowels:
            count += 1

        print("Number of vowels:", count)

```
