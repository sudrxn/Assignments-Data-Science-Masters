```python
ans2:- In Python, exceptions are organized in a hierarchy. The built-in exceptions are derived from the
base class BaseException
```

```python
def print_exception_hierarchy(exception_class, indent=0):
    print('  ' * indent + exception_class.__name__)
    for subclass in exception_class.__subclasses__():
        print_exception_hierarchy(subclass, indent + 1)

if __name__ == "__main__":
    print("Python Exception Hierarchy:")
    print_exception_hierarchy(BaseException)
```

```python
ans3:-In Python, the ArithmeticError class is the base class for exceptions that occur during arithmetic
operations. This class itself is not meant to be instantiated; instead, it serves as a general category for
arithmetic-related exceptions. Some common exceptions derived from ArithmeticError include ZeroDivisionError
,OverflowError, and FloatingPointError
```

```python
#zero division error:- 
# Attempting to divide by zero
try:
    result = 5 / 0
except ZeroDivisionError as e:
    print(f"Error: {e}")
```

```python
#over flow error:-
# Attempting to raise a number to a large power

try:
    result = 10**1
except OverflowError as e:
    print(f"Error: {e}")
```

```python
ans4 = In Python, LookupError is a base class for exceptions that occur when a key or index is not found in
a mapping or sequence, respectively. It is a subclass of the Exception class and serves as a general
category for lookup-related errors
```

```python
#key error :
my_dict = {'a': 1, 'b': 2, 'c': 3}

try:
    value = my_dict['x']  # 'x' is not a key in the dictionary
except KeyError as e:
    print(f"KeyError: {e}")
```

```python
#index error :
my_list = [1, 2, 3, 4, 5]

try:
    value = my_list[10]  # Index 10 is out of range for the list
except IndexError as e:
    print(f"IndexError: {e}")
```

```python
ans 5:- In Python, ImportError is a base class for exceptions that occur when there is an issue with 
importing a module. It can be raised for various reasons, such as when the specified module cannot be found,
has syntax errors, or encounters other problems during the import process.
```

```python
#import error:
try:
    import non_existent_module  # Trying to import a module that doesn't exist
except ImportError as e:
    print(f"ImportError: {e}")
```

```python
#module not found error:
try:
    import non_existent_module  # Trying to import a module that doesn't exist
except ModuleNotFoundError as e:
    print(f"ModuleNotFoundError: {e}")
```
