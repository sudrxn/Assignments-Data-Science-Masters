```python
#ans3:

type 1 Error:

Definition: Type 1 error occurs when the null hypothesis (H0) is incorrectly rejected when it is actually
true. In other words, it is a false positive or the mistake of claiming an effect or relationship when 
there isn't one.

Example Scenario:

Suppose a pharmaceutical company is testing a new drug to determine if it's effective in treating a
particular condition. The null hypothesis (H0) is that the drug has no effect, and the alternative 
hypothesis (H1) is that the drug is effective.
Type 1 Error: The company incorrectly concludes that the drug is effective (rejects the null hypothesis)
when, in fact, it has no therapeutic effect. This could lead to the drug being released to the market when
it's not truly beneficial.
Type 2 Error:

Definition: Type 2 error occurs when the null hypothesis (H0) is not rejected when it is actually false. 
In other words, it is a false negative or the mistake of failing to identify a real effect or
relationship.

Example Scenario:

Continuing with the pharmaceutical example, let's say the drug is genuinely effective, but the study
fails to detect this effect.
Type 2 Error: The company fails to reject the null hypothesis
(fails to conclude that the drug is effective) when, in reality, the drug has therapeutic benefits. 
This could result in the rejection or shelving of a potentially useful medication
```

```python
#ans4:

 Bayes' theorem is a mathematical formula used to update the probability of an event based on new
    evidence. In a medical context:

\[ P(\text{Disease | Test}) = \frac{P(\text{Test | Disease}) \cdot P(\text{Disease})}{P(\text{Test})} \]

- \( P(\text{Disease | Test}) \): Probability of having the disease given a positive test result.
- \( P(\text{Test | Disease}) \): Probability of a positive test result given the presence of the disease.
- \( P(\text{Disease}) \): Prior probability of having the disease.
- \( P(\text{Test}) \): Overall probability of a positive test result.

Bayes' theorem helps in updating our initial belief (prior probability) based on the likelihood of the 
observed evidence (test result) given certain conditions.
```

```python
#ans6:



**Problem:**
Suppose you have a deck of cards, and you draw a card at random. Let A be the event that the card drawn 
is a King, and B be the event that the card is a face card (King, Queen, or Jack). You know that there 
are four Kings in the deck, and there are a total of 12 face cards. What is the probability that the
drawn card is a King given that it is a face card?

**Solution:**
1. Identify the probabilities:
   - \( P(A) \): The probability of drawing a King (prior probability) = \( \frac{4}{52} \) 
(since there are 4 Kings in a deck of 52 cards).
   - \( P(B) \): The probability of drawing a face card (prior probability) = \( \frac{12}{52} \) 
    (since there are 12 face cards in a deck of 52 cards).
   - \( P(B|A) \): The probability of drawing a face card given that it is a King = 1
(since all Kings are face cards).

2. Plug these values into Bayes' Theorem:
   \[ P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} \]
   \[ P(A|B) = \frac{1 \cdot \frac{4}{52}}{\frac{12}{52}} \]

3. Simplify the expression:
   \[ P(A|B) = \frac{\frac{4}{52}}{\frac{12}{52}} \]
   \[ P(A|B) = \frac{4}{12} \]
   \[ P(A|B) = \frac{1}{3} \]

Therefore, the probability of drawing a King given that the drawn card is a face card is 
\( \frac{1}{3} \).
```

```python
#ans8:

The margin of error (MOE) in a confidence interval represents the range within which we can reasonably 
expect the true population parameter to fall. It is influenced by factors such as the level of confidence
(e.g., 95% confidence), the variability of the data, and the sample size.

The formula for the margin of error in a confidence interval is typically given by:

\[ MOE = Z \times \left( \frac{\sigma}{\sqrt{n}} \right) \]

where:
- \( Z \) is the Z-score corresponding to the desired confidence level.
- \( \sigma \) is the standard deviation of the population.
- \( n \) is the sample size.

As the sample size (\( n \)) increases, the margin of error decreases. In other words, a larger sample
size provides more information about the population, resulting in a more precise estimate of the parameter. This relationship is due to the inverse square root relationship in the formula.

For example, let's consider a scenario where you want to estimate the average height of adult males in
a city with 95% confidence. If you have a small sample size, say 50 individuals, your margin of error 
might be relatively large. However, if you increase the sample size to 500 individuals while keeping 
other factors constant, the margin of error would decrease. The larger sample size allows for a more
accurate estimation of the population parameter, leading to a smaller margin of error and a more precise 
confidence interval.
```

```python
#ans9:

The z-score (or standard score) for a data point in a normal distribution is calculated using the 
formula:

\[ Z = \frac{{X - \mu}}{{\sigma}} \]

where:
- \( Z \) is the z-score,
- \( X \) is the raw score,
- \( \mu \) is the population mean, and
- \( \sigma \) is the population standard deviation.

In your case:
- \( X = 75 \) (the data point),
- \( \mu = 70 \) (the population mean), and
- \( \sigma = 5 \) (the population standard deviation).

\[ Z = \frac{{75 - 70}}{{5}} = 1 \]

So, the z-score for a data point of 75 in a distribution with a mean of 70 and a standard deviation
of 5 is \( Z = 1 \).

Interpretation:
A z-score of 1 means that the data point (75) is 1 standard deviation above the mean. In other words, 
it is relatively higher than the average in the population. The positive sign indicates that it is above
the mean. The magnitude of the z-score (1 in this case) gives you an idea of how far the data point 
is from the mean in terms of standard deviations.
```

```python
#ans12:


To conduct a hypothesis test to determine if there is a significant difference in student performance
between two teaching methods (Sample A and Sample B), you can use a two-sample t-test. Here are the steps:

**Hypotheses:**
- Null hypothesis (H0): There is no significant difference in the means of the two teaching methods.
(μA - μB = 0)
- Alternative hypothesis (H1): There is a significant difference in the means of the two teaching methods.
(μA - μB ≠ 0)

**Significance level (α):** 0.01

**Given data:**
- Sample A mean (x̄A) = 85
- Sample A standard deviation (sA) = 6
- Sample A size (nA) is not given, assuming the sample size is reasonably large.
- Sample B mean (x̄B) = 82
- Sample B standard deviation (sB) = 5
- Sample B size (nB) is not given, assuming the sample size is reasonably large.

**Assumptions:**
- Assuming both samples are independent and come from normal distributions.

**Test statistic:**
\[ t = \frac{{(\bar{x}_A - \bar{x}_B)}}{{\sqrt{\left(\frac{{s_A^2}}{{n_A}}\right)
                                               + \left(\frac{{s_B^2}}{{n_B}}\right)}}} \]

**Degrees of freedom:**
\[ df = \frac{{\left(\frac{{s_A^2}}{{n_A}}
                     + \frac{{s_B^2}}{{n_B}}\right)^2}}{{\frac{{\left(\frac{{s_A^2}}{{n_A}}\right)^2}}{{n_A - 1}} 
                    + \frac{{\left(\frac{{s_B^2}}{{n_B}}\right)^2}}{{n_B - 1}}}} \]

**Decision Rule:**
- If the absolute value of the calculated t-statistic is greater than the critical t-value for a
two-tailed test at a significance level of 0.01 and the degrees of freedom, reject the null hypothesis.

**Calculation:**
Plug in the given values into the formula for the t-statistic and calculate the degrees of freedom.
Then compare the absolute value of the t-statistic to the critical t-value.

**Conclusion:**
- If the absolute value of the t-statistic is greater than the critical t-value, reject the null
hypothesis. Otherwise, fail to reject the null hypothesis.

Note: If the sample sizes are large, the t-distribution approaches the normal distribution, 
and you can use a standard normal distribution z-test instead.
```
