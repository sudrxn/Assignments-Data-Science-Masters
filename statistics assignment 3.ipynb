```python
#ans3:

import math

def normal_pdf(x, mean, std_dev):
    """
    Calculate the probability density function (PDF) of a normal distribution.

    Parameters:
    - x (float): The point at which to calculate the PDF.
    - mean (float): The mean of the normal distribution.
    - std_dev (float): The standard deviation of the normal distribution.

    Returns:
    float: The PDF value at the given point.
    """
    if std_dev <= 0:
        raise ValueError("Standard deviation must be greater than 0.")

    coefficient = 1 / (math.sqrt(2 * math.pi) * std_dev)
    exponent_term = math.exp(-((x - mean) ** 2) / (2 * std_dev ** 2))

    pdf = coefficient * exponent_term
    return pdf

# Example usage:
mean = 0
std_dev = 1
x_value = 1.5
pdf_value = normal_pdf(x_value, mean, std_dev)

print(f"The PDF at x={x_value} for N({mean}, {std_dev}) is: {pdf_value}")
```

```python
#ans4:


The binomial distribution is a discrete probability distribution that describes the number of successes in 
a fixed number of independent Bernoulli trials, where each trial has only two possible outcomes: success
or failure. The key properties of the binomial distribution include:

1. **Two Outcomes:** Each trial has only two possible outcomes, typically labeled as success (S) and
failure (F).

2. **Fixed Number of Trials (n):** The number of trials is fixed in advance.

3. **Independence:** The trials are independent, meaning the outcome of one trial does not affect the
outcome of another.

4. **Constant Probability (p):** The probability of success (denoted as p) is constant for each trial.

5. **Discreteness:** The random variable representing the number of successes is a discrete variable,
taking on non-negative integer values.

6. **Probability Mass Function (PMF):** The probability mass function of the binomial distribution is
given by the formula:

   \[ P(X = k) = \binom{n}{k} p^k (1-p)^{n-k} \]

   where \( n \) is the number of trials, \( k \) is the number of successes, \( p \) is the probability
    of success on a single trial, and \( (1-p) \) is the probability of failure on a single trial.

Two examples of events where the binomial distribution can be applied:

1. **Coin Flipping:**
   Consider flipping a fair coin (where the probability of getting heads, for example, is \( p = 0.5 \)). 
    If you flip the coin 10 times, the number of heads you get follows a binomial distribution with
    parameters \( n = 10 \) (number of trials) and \( p = 0.5 \) (probability of success).

2. **Quality Control:**
   In manufacturing, the production of items often involves testing for defects. Suppose a quality
    control inspector examines a random sample of 20 items and classifies each as either defective or 
    non-defective. If the probability of a defective item is \( p = 0.1 \), then the number of defective
    items in the sample follows a binomial distribution with parameters \( n = 20 \) (number of trials)
    and \( p = 0.1 \) (probability of a defective item).
```

```python
#ans6:


from math import exp, factorial

def poisson_cdf(mean, k):
    """
    Calculate the cumulative distribution function (CDF) of a Poisson distribution.

    Parameters:
    - mean: The mean of the Poisson distribution.
    - k: The point at which to calculate the CDF.

    Returns:
    The CDF value at the given point.
    """
    if mean < 0 or k < 0:
        raise ValueError("Mean and k must be non-negative.")

    cdf = 0.0
    for i in range(k + 1):
        cdf += exp(-mean) * (mean ** i) / factorial(i)

    return cdf

# Example usage:
mean_value = 2.5
point_value = 3
result = poisson_cdf(mean_value, point_value)
print(f"P(X <= {point_value}) for Poisson(mean={mean_value}): {result}")
```

```python
#ans7:

The binomial distribution and the Poisson distribution are both probability distributions that describe
the number of events or successes in a fixed number of trials or a fixed interval. However, there are 
key differences between the two distributions.

#Nature of the Process:

Binomial Distribution: It is used to model the number of successes (or failures) in a fixed number of
independent and identical trials, where each trial has only two possible outcomes (success or failure).

Poisson Distribution: It is used to model the number of events that occur in a fixed interval of time or
space. It is often applied when the events are rare and the probability of success in each small 
subinterval is small.

#Number of Possible Outcomes:

Binomial Distribution: Each trial in a binomial distribution has two possible outcomes (success or failure).

Poisson Distribution: It can represent an infinite number of possible outcomes (events) in the specified
interval.

#Probability of Success:

Binomial Distribution: The probability of success remains constant across all trials.

Poisson Distribution: The probability of an event occurring in a small subinterval is proportional to the
length of that subinterval, but it does not necessarily remain constant across all subintervals.



#Applicability:

Binomial Distribution: It is suitable when there is a fixed number of trials, and each trial has two 
possible outcomes.

Poisson Distribution: It is suitable for modeling rare events occurring in a fixed interval.

#Shape of the Distribution:

Binomial Distribution: When the number of trials is large and the probability of success is small, 
the binomial distribution approaches a normal distribution.

Poisson Distribution: It is a skewed distribution, especially when the event rate is low.


```

```python
#ans9:


The mean (or expected value) and variance are related in both the Binomial and Poisson distributions,
but the specific relationships differ for each distribution.

1. **Binomial Distribution:**
   - **Mean (μ):** For a binomial distribution with parameters n (number of trials) and p 
    (probability of success on each trial), the mean is given by μ = np.
   - **Variance (σ²):** The variance of a binomial distribution is given by σ² = np(1 - p).

   In summary, for a binomial distribution, both the mean and variance depend on the number of trials (n) 
    and the probability of success on each trial (p).

2. **Poisson Distribution:**
   - **Mean (μ):** For a Poisson distribution, the mean is denoted by λ (lambda). The mean is also
    equal to the variance, so μ = σ² = λ.

   In summary, for a Poisson distribution, the mean and variance are equal and are both represented by
the parameter λ.

In contrast, for many other distributions, the mean and variance may not necessarily be equal. The 
relationships described for the Binomial and Poisson distributions are specific to these distributions 
and are a result of their mathematical properties.
```
