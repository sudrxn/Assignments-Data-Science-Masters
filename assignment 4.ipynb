```python
Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.


ans: A class in object-oriented programming (OOP) is a blueprint or template that specifies the properties and characteristics
     of a particular kind of object. On the other hand, an object is an instance of a class with its own particular set of values
      for the class's defined attributes.

      Let's use the class "Car" as an example.
      All automobiles have some characteristics, including the make, model, and year, as well as specific actions, like accelerating,
      braking, and turning. These characteristics and actions would be defined by the "Car" class. 
    
    #example:=
    
    class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.speed = 0
        
    def accelerate(self):
        self.speed += 10
        
    def brake(self):
        self.speed -= 10
        
    def turn(self, direction):
        print("Turning", direction)
        
        
        
        
        
        
        
        
        
Q2. Name the four pillars of OOPs.

ans: Object-Oriented Programming's (OOPs) four pillars are:

   1) Encapsulation: This is the process of combining data and the methods (functions) that work on it into a single entity,
     like a class. Encapsulation makes a programme more modular and easier to maintain by allowing data to be hidden from other
     programme components and accessed only through a specific interface.
    
  2)Abstraction: This involves identifying the essential characteristics of an object or system and ignoring all others.
    Abstraction allows complex systems to be broken down into simpler parts, which can be independently designed and implemented.
    It also allows the implementation of an object to be modified without affecting its external behavior.
    
  3)Inheritance: This is a mechanism by which one class can inherit the properties and behaviors of another class.
    Inheritance allows new  classes to be built on top of existing ones, reducing redundancy and improving code reuse.  
    
  4)Polymorphism, which translates as "many forms," describes an object's capacity to adopt a variety of distinct shapes 
    depending on the situation in which it is used. Method calling and method overloading, which enable various objects to 
    respond differently to the same message or method call, are frequently used to create polymorphism.  

    
    
    
    
    
    
    
    
    
Q3. Explain why the __init__() function is used. Give a suitable example.

ans: When a class instance is created in Python, the __init__() function is a unique method that is called. 
     This function is used to set up any additional setup requirements as well as initialise the object's attributes.

     The initial state of an object is frequently defined using the __init__() function. If you have a Vehicle class,
        for instance, you could use the __init__() method to establish the make, model, and year attributes' initial values.
        
        
       #example:-
    
    class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def get_description(self):
        return self.make,self.model,self.year
    
    
    
    
    
    
    
    
    
    
    
Q4. Why self is used in OOPs?

ans: "Self" is a term used in object-oriented programming (OOP) to denote the instance of the class that is presently being worked
     on. The term "Self" refers to the object to which the method or attribute is a part.

    Every element in OOP is an object with its own set of characteristics and methods. A function or attribute can refer to the 
    specific object it is a part of by using the term "self" rather than just a generic class.
    
    
    Let's use the class "Person" as an example. It has a function called "greet." We can use the "greet" function to interact 
    with a Person class instance that we have created. The "self" keyword allows us to refer to the particular instance of
    the Person class that the "greet" method is being called on.
    
    
    #example
    
    class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print("Hello, my name is", self.name)

    # create an instance of the Person class
    person = Person("John")

    # call the greet method on the person instance
    person.greet()

    
    
    
    
    
    
    
    
Q5. What is inheritance? Give an example for each type of inheritance.

ans: A key idea in object-oriented programming (OOP) is inheritance, which enables a class to take on traits and functions 
    from other classes. The class being inherited from is known as the superclass or base class, while the class being inherited
    from is known as the subclass or derived class. The ability to create complex class hierarchies is made possible via inheritance,
    which promotes code reuse.

    #Four types of inheritance exist:
    
    1) Single inheritance: In single inheritance, a subclass inherits from a single superclass. The subclass inherits all the
       properties and methods of the superclass.
        
        #for example
        
        class Animal:
    def move(self):
        print("I can move.")

    class Dog(Animal):
    def bark(self):
        print("I can bark.")

    d = Dog()
    d.move()  # Output: I can move.
    d.bark()  # Output: I can bark.

    
   2)  Multiple inheritance: In multiple inheritance, a subclass inherits from multiple superclasses. The subclass inherits
       all the properties and methods of all the superclasses. 
       
        #For example:
        
        class A:
    def method_a(self):
        print("method_a")

    class B:
    def method_b(self):
        print("method_b")

    class C(A, B):
    def method_c(self):
        print("method_c")

    c = C()
    c.method_a()  # Output: method_a
    c.method_b()  # Output: method_b
    c.method_c()  # Output: method_c

    
    
    3) Multi-level inheritance: In multi-level inheritance, a subclass inherits from a superclass, which in turn inherits 
      from another superclass. 
        
        #For example:
        
        class Animal:
    def move(self):
        print("I can move.")

    class Dog(Animal):
    def bark(self):
        print("I can bark.")

    class Bulldog(Dog):
    def run(self):
        print("I can run.")

    b = Bulldog()
    b.move()  # Output: I can move.
    b.bark()  # Output: I can bark.
    b.run()   # Output: I can run.

    
    
    4) Hierarchical inheritance: In hierarchical inheritance, multiple subclasses inherit from a single superclass. 
    
    #For example:
    
    class Animal:
    def move(self):
        print("I can move.")

    class Dog(Animal):
    def bark(self):
        print("I can bark.")

    class Cat(Animal):
    def meow(self):
        print("I can meow.")

    d = Dog()
    d.move()  # Output: I can move.
    d.bark()  # Output: I can bark.

    c = Cat()
    c.move()  # Output: I can move.
    c.meow()  # Output: I can meow.


      
```
