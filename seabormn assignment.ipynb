```python
#ans1 : 

Scatter Plot - Visualize the relationship between two continuous variables.
Line Plot - Show the trend of a continuous variable over a specific range.
Bar Plot - Display the distribution of categorical data or compare quantities of different categories.
Histogram - Visualize the distribution of a single continuous variable.
Box Plot - Display the distribution of a dataset and identify outliers.
```

```python
#ans2:- 

import seaborn as sns
fmri= sns.load_dataset("fmri")
fmri
sns.lineplot(x =fmri.timepoint ,y =fmri.signal, data=fmri)
```

```python
#ans3:using x ='pclass', y = 'age'
titanic= sns.load_dataset("titanic")
sns.boxplot(x =titanic.pclass, y =titanic.age, data=titanic)
```

```python
#using x ='pclass', y = 'fare'
sns.boxplot(x =titanic.pclass, y =titanic.fare, data=titanic)
```

```python
diamond= sns.load_dataset("diamonds")
sns.histplot(x=diamond.price,hue="cut", data=diamond)
diamond
```

```python
diamond["price"].value_counts()
```

```python
iris=sns.load_dataset("iris")
sns.pairplot(hue="species",data=iris)
```

```python
flight=sns.load_dataset("flights")
flight_pivot=flight.pivot("year","month","passengers")
sns.heatmap(flight_pivot, annot=True, fmt="d",  linewidths=.5)
```
