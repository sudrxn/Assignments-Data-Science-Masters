```python
#ans1:

Probability Mass Function (PMF) and Probability Density Function (PDF) are mathematical concepts used in 
probability theory and statistics to describe the likelihood of different outcomes in a random experiment.

1. **Probability Mass Function (PMF):**
   - The PMF is used for discrete random variables, which take on a finite or countably infinite set
    of distinct values.
   - It gives the probability that a discrete random variable is exactly equal to a particular value.
   - Mathematically, for a discrete random variable X, the PMF is denoted as P(X = x), where x is a
    specific value that X can take.

   **Example:**
   Consider a fair six-sided die. The random variable X represents the outcome of a single roll. The PMF
    for X is given by:

   \[ P(X = 1) = P(X = 2) = P(X = 3) = P(X = 4) = P(X = 5) = P(X = 6) = \frac{1}{6} \]

   Here, the PMF assigns equal probabilities (1/6) to each possible outcome.

2. **Probability Density Function (PDF):**
   - The PDF is used for continuous random variables, which can take on any value within a range.
   - Instead of giving the probability at a specific point, the PDF provides the probability density over
a range of values.
   - The probability of a continuous random variable falling within a specific interval is given by the
    area under the PDF curve over that interval.

   **Example:**
   Let's consider a continuous random variable Y representing the height of individuals in a population. 
    The PDF for Y might be a normal distribution (bell curve). The PDF, denoted as f(y), provides the 
    probability density for different heights.

   \[ f(y) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(y - \mu)^2}{2\sigma^2}} \]

   Here, \(\mu\) is the mean height, \(\sigma\) is the standard deviation, and \(e\) is the base of
    the natural logarithm. The probability of an individual having a height within a specific range
    \([a, b]\) is given by the integral of the PDF over that range:

   \[ P(a \leq Y \leq b) = \int_{a}^{b} f(y) \,dy \]

   The area under the curve within the interval \([a, b]\) represents the probability of the random 
    variable falling within that range.

In summary, PMF is used for discrete random variables, providing the probability at specific points,
while PDF is used for continuous random variables, giving the probability density over ranges of values.
```

```python
#ans2:

A Cumulative Density Function (CDF) is a mathematical function that describes the probability that a random 
variable takes on a value less than or equal to a given point. In other words, it gives the cumulative
probability distribution of a random variable. The CDF is often denoted by the letter F and is defined as:

\[ F(x) = P(X \leq x) \]

Here, \(X\) is the random variable, and \(F(x)\) represents the probability that \(X\) takes a value less 
than or equal to \(x\).

The CDF provides a way to understand the distribution of a random variable and is particularly useful in
probability and statistics. It has several important properties:

1. **Non-decreasing:** The CDF is a non-decreasing function, meaning that as the value of \(x\) increases,
the probability \(P(X \leq x)\) does not decrease.

2. **Limits:** The limit of the CDF as \(x\) approaches negative infinity is 0, and as \(x\) approaches 
positive infinity, the limit is 1.

3. **Right-Continuity:** The CDF is a right-continuous function, meaning that it has no jumps; the
probability at a specific point is the limit from the right.

Here's an example to illustrate the concept of CDF:

Suppose you have a six-sided die. Let \(X\) be the random variable representing the outcome of a single 
roll. The CDF for this die would be:

\[ F(x) = \frac{1}{6} \sum_{i=1}^{x} i \]
where \(x\) is the outcome of the die roll. This CDF essentially tells you the probability of getting a
number less than or equal to \(x\) on a single die roll.

For instance, if you want to find the probability of getting a 4 or less, you would evaluate \(F(4)\):

\[ F(4) = \frac{1}{6} \sum_{i=1}^{4} i = \frac{1}{6} \times (1 + 2 + 3 + 4) = \frac{10}{36} \]

CDFs are used in statistics for various purposes, including determining percentiles, calculating 
probabilities, and understanding the overall distribution of a random variable. They provide a 
comprehensive view of the probability distribution and are an essential tool in statistical analysis.
```

```python
#ans3

The normal distribution, also known as the Gaussian distribution or bell curve, is a widely used probability distribution in various fields due to its mathematical properties and its tendency to describe many natural phenomena. Here are some examples of situations where the normal distribution might be used as a model:

1. **Biological Measurements:**
   - Height and weight of individuals in a population.
   - Blood pressure readings.

2. **Economics and Finance:**
   - Stock prices.
   - Returns on investments.
   - Income distribution.

3. **Psychometrics:**
   - IQ scores.
   - Test scores in a population.

4. **Natural Phenomena:**
   - Measurement errors.
   - Meteorological data, such as temperature.

5. **Quality Control:**
   - Product dimensions in manufacturing.
   - Defects in a production process.

6. **Physics:**
   - Measurement errors in experimental data.
   - Particle velocities in gas.

7. **Social Sciences:**
   - Distribution of scores on standardized tests.
   - Human abilities and traits.

Parameters of the normal distribution include:

1. **Mean (μ):** The central location of the distribution. It represents the average or expected value.

2. **Standard Deviation (σ):** The measure of the spread or dispersion of the distribution. A larger
standard deviation implies a wider and flatter curve, while a smaller standard deviation results in a 
narrower and taller curve.

The relationship between these parameters and the shape of the normal distribution can be summarized as
follows:

- **Mean (μ):** Shifts the entire distribution left or right along the horizontal axis. A higher mean moves
the distribution to the right, and a lower mean moves it to the left.

- **Standard Deviation (σ):** Controls the spread of the distribution. A larger standard deviation leads to
a wider and flatter curve, while a smaller standard deviation results in a narrower and taller curve.

In summary, the mean determines the center, and the standard deviation determines the spread of the normal
distribution. Together, they provide a concise description of the shape and characteristics of the
distribution.
```
